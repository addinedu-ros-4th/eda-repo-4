```python
import seaborn as sns
import matplotlib.pyplot as plt 
import koreanize_matplotlib
```

```python
sns.catplot(kind='bar',x='키워드_x',y='매출액1',data=result_df, palette='Set3')
sns.catplot(kind='bar', x='키워드_x', y='평균연봉', data=result_df, palette='Set3')
```

```python
# '기업이름' 열을 제외하고 상관관계 행렬 계산
correlation_matrix = result_df.drop(columns=['기업이름','기업규모','키워드_x'], axis=1).corr()

# Heatmap 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='Reds', fmt='.2f', linewidths=.5, )
plt.title('상관관계 Heatmap')
plt.show()
```
