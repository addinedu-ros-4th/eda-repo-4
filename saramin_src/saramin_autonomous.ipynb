```python
driver = webdriver.Chrome(service=Service("../drive/chromedriver-linux64/chromedriver"))

company_name_list = []
category_list = []
job_link_list = []
job_condition_list = []

for i in tqdm(range(1,3),total=):
         
    url = f"https://www.saramin.co.kr/zf_user/jobs/list/job-category?cat_kewd=846&panel_type=&search_optional_item=n&search_done=y&panel_count=y&preview=y&page={i}&page_count=100"
    driver.get(url)
    job_box_1 = driver.find_element(By.CLASS_NAME, 'list_body')
    job_box_2 = job_box_1.find_elements(By.CLASS_NAME, 'box_item')
    time.sleep(1)

    for list in job_box_2:
        time.sleep(1)
        company_name = list.find_element(By.CLASS_NAME, 'col.company_nm').text

        company_name = company_name.split('\n')[0]
        
        company_name_list.append(company_name)
        

        job_condition = list.find_element(By.CLASS_NAME, 'career').text

        job_condition_list.append(job_condition)

        category_list.append('자율주행')
    
        job_link_box = list.find_element(By.CLASS_NAME, 'job_tit')
        job_link_info = job_link_box.find_element(By.TAG_NAME, 'a')
        job_link = job_link_info.get_attribute('href')
        job_link_list.append(job_link)

        print(company_name_list, job_condition_list,  category_list, job_link_list)


```

```python
from bs4 import BeautifulSoup
import requests
import os

def scrape_webpage(soup):
    # 케이스 1: 'table' 태그와 특정 클래스를 가진 요소 탐색
    major_info_box = soup.find('table', class_='cont_recruit_template')
    if major_info_box:
        major_info_box2 = major_info_box.find('table', class_='wrap_tbl_template')
        table_text = major_info_box2.get_text(separator=' ', strip=True)
        return table_text

    # 케이스 2: 'body' 태그와 특정 스타일을 가진 요소 탐색
    major_info_box = soup.find('body', style='margin:0 !important;padding:0 !important;background:#fff !important;')
    if major_info_box:
        major_info_box2 = major_info_box.find('div', class_='user_content')
        return major_info_box2.get_text(separator=' ', strip=True)

    # 두 케이스 모두 실패한 경우 None 반환
    return None

def download_image(image_element, index):
    image_url = image_element['src']
    response = requests.get(image_url)
    if response.status_code == 200:
        save_path = '/home/ryu/amr_ws/EDA/data/job_image'
        os.makedirs(save_path, exist_ok=True)  # 폴더가 없으면 생성
        file_name = f"{index}_{os.path.basename(image_url)}"
        full_path = os.path.join(save_path, file_name)
        with open(full_path, 'wb') as file:
            file.write(response.content)
        return full_path
    return Noneget
```
