### 1차 크롤링

```python
company_list = []

for i in range(1,3):
    url = "https://job.incruit.com/jobdb_list/searchjob.asp?kw=%B5%E5%B7%D0&rgn2=18&rgn2=11&page={page_number}"        # 자신의 찾으려는 웹사이트 입력
    req = Request(url.format(page_number = urllib.parse.quote(str(i))))
    res = urlopen(req)
    soup = BeautifulSoup(res, "html.parser")
    
    element = soup.find_all('ul',class_='c_row')
    for each in element:
        cl_top = each.find('div',class_='cl_top')
        cell_mid = each.find('div',class_='cell_mid')

        company = []
        name =each.find('a',class_='cpname').string    # 회사이름
        title = cell_mid.find(class_='cl_top').get_text()   # 모집명
        link = cell_mid.find('a')['href']                  # 링크
        info = cell_mid.find(class_='cl_md').get_text()     # 경력 학력 주소 연봉
        field = cell_mid.find(class_='cl_btm').get_text()  # 분야

        company.append(name)
        company.append(title)
        company.append(link)
        company.append(info)
        company.append(field)
        company_list.append(company)
```

```python
import pandas as pd 
drone_df = pd.DataFrame(company_list, columns=['기업이름', '모집명','링크','info','담당업무'])
```

### 2차 크롤링

```python
from tqdm import tqdm 
from user_agent import generate_user_agent

address_list = []
link2_list = []

for idx, row in tqdm(drone_df.iterrows(), total=len(drone_df.index)):
    req = Request(url=row['링크'], headers={'User-Agent':generate_user_agent()})
    page = urlopen(req)
    soup = BeautifulSoup(page, 'html.parser')

    try:
        company = soup.find(class_='jcinfo_list last')
        address = company.get_text()
        address_list.append(address)
    except (AttributeError, IndexError):
        address_list.append(None)

    try:
        link = soup.find_all('div',class_='section_view_layout')[1]
        link2 = link.find('iframe',id='ifrmJobCont')['src']
        link2_list.append(link2)
    except (AttributeError, IndexError):
        link2_list.append(None)
```

```python
address_list, link2_list
```

### 데이터 삽입

```python
drone_df['주소'] = address_list
drone_df['링크2'] = link2_list
```

```python
import re

# 주소 열에서 주소라는 단어를 발견했을 때 이후의 문자열을 추출하는 함수
def extract_address(text):
    if isinstance(text, str):  # 문자열인 경우에만 처리
        match = re.search(r'주소\s*(.*)', text)
        return match.group(1) if match else None
    else:
        return None

# 주소 열에 적용하여 새로운 열 생성
drone_df['추출된_주소'] = drone_df['주소'].apply(extract_address)

# 결과 확인
print(drone_df[['주소', '추출된_주소']])
```

### 불용어 처리

```python
from nltk.tokenize import word_tokenize
from nltk import FreqDist

# Required List에서 토큰화하고 불용어(stop words) 제거
stop_words = set(['•', '[', ']', '(', ')', ',','-', '및', '또는','혹은', '관련', '분', '등','대한','이상',':', '\n','*','채용','인크루트',
                  '자격요건','>','@','주','수','지원','전형단계','접수방법','근무부서','모집부문','있습니다', '후', '근무조건','서류전형',
                  '제출서류','명','경우','이력서','/','?','담당업무','.','--','연봉','우대사항','입사지원서','개발','!','고용형태','급여조건',
                  '유의사항','취소될','있는','허위사실이','있을','비희망','기타','근무','정규직','0','협의','채용이','직급/직책','최종합격추가',
                  '자기소개서','~','최종합격추가','면접','관리','가능','경력사항','5일','채용시','업무','면접진행','결정','인원',re.compile(r'.*년.*')
                  ,'제출서류이력서','근무시간','오전','00','%','함께','오후','수습기간3개월','기획','채용시스템접수양식','신입/경력','우대',
                  '2','경험자','포함','시','이메일','장애인채용희망여부','주소정근로시간','이메일','포함','프로그램','상여금','최종심사',
                  '주소정근로시간','미','월','필수','학력사항','이','제출서류에','회사로3D스캐너','영업팀급여조건','소개해주세요','가능한분','경력사항필히포함',
                  '모집합니다','업무는',''])

keywords = []

for text in drone_df['필요역량']:
    tokens = word_tokenize(text)
    tokens = [token for token in tokens if token.lower() not in stop_words]
    keywords.extend(tokens)

# 빈도 분포 계산
freq_dist = FreqDist(keywords)

# 가장 많이 나오는 상위 30개 키워드 출력
top_keywords = freq_dist.most_common(30)
```

```python
import re

# 주소에서 (숫자) 제거하는 함수
def remove_numbers_in_parentheses(address):
    if pd.isna(address):
        return address
    return re.sub(r'\(\d+\)', '', address)

# '주소' 열에 함수 적용
drone_df['추출된_주소'] = drone_df['추출된_주소'].apply(remove_numbers_in_parentheses)

```
