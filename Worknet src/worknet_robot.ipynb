```python
# 다음 페이지 넘어가기(테스트)
next_page = driver.find_element(By.CSS_SELECTOR, '#mForm > div.nav_wrp > nav > a:nth-child(2)')
next_page.send_keys(Keys.ENTER)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By 
from selenium.webdriver.common.keys import Keys 
import pandas as pd 
import time

name = []
kind_of = []
salary = []
response = []
education = []
add = []

for j in range(1,6):
    # 다음 페이지 넘어가기
     next_page = driver.find_element(By.CSS_SELECTOR, '#mForm > div.nav_wrp > nav > a:nth-child({})'.format(j))
     next_page.send_keys(Keys.ENTER)
    
     body1 = driver.find_element(By.CSS_SELECTOR, '#mForm > div.table-wrap > table > tbody')

     for i in range(1,11):
         selector_name = f'#list{i} > td:nth-child(2) > a'
         selector_kind_of = f'#list{i} > td:nth-child(3) > div > p:nth-child(3) > em:nth-child(1)'
         selector_salary = f'#list{i} > td:nth-child(4) > div > p:nth-child(1) > strong'
         selector_response = f'#jobContLine{i}'
         selector_education = f'#list{i} > td:nth-child(3) > div > p:nth-child(3) > em:nth-child(2)'
         selector_add = f'#list{i} > td:nth-child(3) > div > p:nth-child(3) > em:nth-child(3)'

             
         name_element = body1.find_element(By.CSS_SELECTOR, selector_name)
         kind_of_element = body1.find_element(By.CSS_SELECTOR, selector_kind_of)
         salary_element = body1.find_element(By.CSS_SELECTOR, selector_salary)
         response_element = body1.find_element(By.CSS_SELECTOR, selector_response)
         education_element = body1.find_element(By.CSS_SELECTOR, selector_education)
         add_element = body1.find_element(By.CSS_SELECTOR, selector_add)

        
         if name_element and kind_of_element and salary_element and response_element and education_element and add_element:
             
             name.append(name_element.text)
             kind_of.append(kind_of_element.text)
             salary.append(salary_element.text)
             response.append(response_element.text)
             education.append(education_element.text)
             add.append(add_element.text)    

             time.sleep(3)
```

### 필요역량 column 재수정 키워드에 맞게 적절하게

```python
from selenium import webdriver
from selenium.webdriver.common.by import By 
import pandas as pd 
import time

prere = []

body1 = driver.find_element(By.CSS_SELECTOR, '#mForm > div.table-wrap > table > tbody')

# 예시 키워드 리스트
keywords = ['정보', '통신', 'IT', '딥러닝', '빅데이터', 'AI', '데이터', '인공지능', '소프트웨어', 'C', 'C++', '임베디드', 'ROS', 'ROS2', 'C#', 'AI', 'JAVA', 'Spring', '머신러닝', 'JSP', '자율주행', '파이썬'
            'PLC','2차전지','Display','엔지니어','HW','SW','회계','기획','사무','Linux','OpenCV','MATLAB','TensorFlow','자동화','CAD','전장','Visual C++','생산','기술']

for i in range(1, 11):
    selector_prere = f'#list{i} > td:nth-child(3) > div > div > a'
             
    prere_element = body1.find_element(By.CSS_SELECTOR, selector_prere)
    
    if any(keyword in prere_element.text for keyword in keywords):
        prere.append(prere_element.text)
    else:
        prere.append('-')

    time.sleep(3)

prere
```
