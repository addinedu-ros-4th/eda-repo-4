```python
from selenium import webdriver
from selenium.webdriver.common.by import By 
import pandas as pd 
import time

prere = []

body1 = driver.find_element(By.CSS_SELECTOR, '#mForm > div.table-wrap > table > tbody')

# 예시 키워드 리스트
keywords = ['정보', '통신', 'IT', '딥러닝', '빅데이터', 'AI', '데이터', '인공지능', '소프트웨어', 'C', 'C++', '임베디드', 'ROS', 'ROS2', 'C#', 'AI', 'JAVA', 'Spring', '머신러닝', 'JSP', '자율주행', '파이썬']

for i in range(1, 11):
    selector_prere = f'#list{i} > td:nth-child(3) > div > div > a'
             
    prere_element = body1.find_element(By.CSS_SELECTOR, selector_prere)
    
    if any(keyword in prere_element.text for keyword in keywords):
        prere.append(prere_element.text)
    else:
        prere.append('-')

    time.sleep(3)

prere
```
