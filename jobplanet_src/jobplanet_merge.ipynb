```python
import pandas as pd
```

```python
df1 = pd.read_csv('../../../data/eda_project_data/잡플래닛/IT.csv')
df2 = pd.read_csv('../../../data/eda_project_data/잡플래닛/건설업.csv')
df3 = pd.read_csv('../../../data/eda_project_data/잡플래닛/금속재료철강.csv')
df4 = pd.read_csv('../../../data/eda_project_data/잡플래닛/기계설비.csv')
df5 = pd.read_csv('../../../data/eda_project_data/잡플래닛/농업임업어업광업.csv')
df6 = pd.read_csv('../../../data/eda_project_data/잡플래닛/반도체디스플레이.csv')
df7 = pd.read_csv('../../../data/eda_project_data/잡플래닛/석유화학에너지.csv')
df8 = pd.read_csv('../../../data/eda_project_data/잡플래닛/전기전자.csv')
df9 = pd.read_csv('../../../data/eda_project_data/잡플래닛/조선항공우주.csv')
```

```python
revenue = pd.read_csv('../../../data/eda_project_data/매출액/rev_merge.csv')
```

```python
revenue['순이익성장지수'] = ((revenue['순이익1']-revenue['순이익3'])/revenue['순이익3'])
revenue['매출액성장지수'] = ((revenue['매출액1']-revenue['매출액3'])/revenue['매출액3'])
revenue['자본총계성장지수'] = ((revenue['자본총계1']-revenue['자본총계3'])/revenue['자본총계3'])
```

```python
revenue.sort_values(by='순이익성장률',ascending=False)
```

```python
revenue_duplicate = revenue.drop_duplicates(subset='기업이름')
```

```python
revenue_duplicate = revenue_duplicate[revenue_duplicate['자본총계1']!=0]
```

```python
revenue_duplicate.reset_index(drop=True,inplace=True)
```

```python
revenue_duplicate.to_csv('../../../data/eda_project_data/매출액/기업성장.csv',index=False)
```
